## Information Value (IV)

```python
df1['IV'] = (df1['prop_n_good']-df1['prop_n_bad']) * df1['WoE']
df1['IV'] = df1['IV'].sum() # information value is the same for all categories( sum of all IV's)
df1
```

                    Information Value	Predictive Power
                               < 0.02	useless for prediction
                          0.02 to 0.1    Weak predictor
                           0.1 to 0.3	Medium predictor
                           0.3 to 0.5	Strong predictor
                                  0.5	Suspicious or too good to be true
 
Grade is a medium predictor for the good/bad variable

```python
# create a function to return WoE and IV for all categorical variables
def woe_cat(df, cat_var, good_bad_var): # dataframe that contains the catetegorical variables, two variables in question
    df=pd.concat([df[cat_var], good_bad_var], axis=1)
    df=pd.concat([df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(), 
                  df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1)
    df=df.iloc[: , [0,1,3]] #locate the three relevant columns
    df.columns=[df.columns.values[0], 'N_obs', 'prop_good'] #rename the columns
    df['prop_N_obs']=df['N_obs'] / df['N_obs'].sum() # predicts how well a category can distinguish good from bad
    # total number and proportons of good and bad obs in each grade
    df['n_good']=df['prop_good'] * df['N_obs']
    df['n_bad']=(1-df['prop_good']) * df['N_obs']
    df['prop_n_good']=df['n_good'] / df['n_good'].sum()
    df['prop_n_bad']=df['n_bad'] / df['n_bad'].sum()
    df['WoE']=np.log(df['prop_n_good'] / df['prop_n_bad']) # weight of evidence for each category
    df=df.sort_values(['WoE']) # category with the highest defaulter comes first
    df=df.reset_index(drop=True)
    # difference of weight of two subsequent categories
    df['diff_prop_good']=df['prop_good'].diff().abs() # absolute value
    df['diff_WoE']=df['WoE'].diff().abs()
    df['IV']=(df['prop_n_good']-df['prop_n_bad']) * df['WoE']
    df['IV']=df['IV'].sum() # information value is the same for all categories( sum of all IV's)
    return df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# visual plots of the weights of evidence
def woe_plot(df, rotation_of_x_axis_labels = 0):
    x=np.array(df.iloc[:, 0].apply(str)) # convert categories to string place then in N-D array for matplotlib plotting 
    y=df['WoE']
    plt.figure(figsize=(18, 6)) #figure size of 18 by 6
    plt.plot(x, y, marker='o', linestyle='--', color='k')
    plt.xlabel(df.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title(str('Weight of Evidence by ' + df.columns[0]))
    plt.xticks(rotation=rotation_of_x_axis_labels)
```
