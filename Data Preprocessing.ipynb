```python
loan_data['total_rev_hi_lim'].fillna(loan_data['funded_amnt'], inplace=True)
```

```python
# replace all missing values with their means
loan_data['annual_inc'].fillna(loan_data['annual_inc'].mean(), inplace=True)
loan_data['mnths_since_earliest_cr_line'].fillna(loan_data['mnths_since_earliest_cr_line'].mean(), inplace=True)
loan_data['acc_now_delinq'].fillna(loan_data['acc_now_delinq'].mean(), inplace=True)
loan_data['total_acc'].fillna(loan_data['total_acc'].mean(), inplace=True)
loan_data['pub_rec'].fillna(loan_data['pub_rec'].mean(), inplace=True)
loan_data['open_acc'].fillna(loan_data['open_acc'].mean(), inplace=True)
loan_data['inq_last_6mths'].fillna(loan_data['inq_last_6mths'].mean(), inplace=True)
loan_data['delinq_2yrs'].fillna(loan_data['delinq_2yrs'].mean(), inplace=True)
loan_data['emp_length_int'].fillna(loan_data['emp_length_int'].mean(), inplace=True)
```

## Data preparation for model training 

```python
# check the dependent variable loan status
loan_data['loan_status'].unique()
# fully paid, current, in grace perod are all good among others
```

```python
# get total of counts for each unque value in loan status
loan_data['loan_status'].value_counts()
```

```python
# value proportion for each count category
loan_data['loan_status'].value_counts() / loan_data['loan_status'].count()
```

```python
# create a variable of only 0 (bad) and 1 (good)
loan_data['good_bad'] = np.where(loan_data['loan_status'].isin(['Charged Off', 'Late (31-120 days)', 
                                                                'Does not meet the credit policy. Status:Charged Off', 
                                                                'Default']), 0, 1)
```

```python
from sklearn.model_selection import train_test_split
trainX, testX, trainY, testY = train_test_split(loan_data.drop('good_bad', axis=1), loan_data['good_bad'], 
                                                test_size=0.2, random_state=42)
```

```python
# check dimensions of the split dataframes
print('trainX: ', trainX.shape, 'trainY: ', trainY.shape, 'testX: ', testX.shape, 'testY: ', testY.shape)
```

```python
trainX.to_csv('trainX.csv', index=None, header=True)
trainY.to_csv('trainY.csv', index=None, header=True)
testX.to_csv('testX.csv', index=None, header=True)
testY.to_csv('testY.csv', index=None, header=True)
```
